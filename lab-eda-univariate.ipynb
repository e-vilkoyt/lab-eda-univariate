```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el dataset
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
# Mostrar las primeras filas para conocer el contenido
df.head()
```

## 1. Exploración de categorías de productos

```python
# Generar una tabla de frecuencias para la columna 'category'
category_freq = df['category'].value_counts()

# Mostrar las 5 categorías más listadas
top_5_categories = category_freq.head(5)
print("Top 5 categorías más listadas:\n", top_5_categories)

# Visualización: Gráfico de barras para todas las categorías
plt.figure(figsize=(10,6))
sns.barplot(x=category_freq.index, y=category_freq.values)
plt.xticks(rotation=90)
plt.title("Distribución de productos por categoría")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia")
plt.show()

# Visualización: Pie chart para las 5 categorías más listadas
top_5_categories.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Proporción de las 5 categorías más listadas")
plt.ylabel("")
plt.show()
```

## 2. Análisis de precios de productos

```python
# Calcular medidas de centralidad
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Media de precios: {mean_price}")
print(f"Mediana de precios: {median_price}")
print(f"Moda de precios: {mode_price}")

# Calcular medidas de dispersión
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Varianza: {variance_price}")
print(f"Desviación estándar: {std_dev_price}")
print(f"Rango de precios: {range_price}")
print(f"Rango intercuartílico: {iqr_price}")

# Visualización: Histograma para la distribución de precios
plt.figure(figsize=(8,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribución de precios de productos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()

# Visualización: Box plot para detectar outliers
plt.figure(figsize=(8,6))
sns.boxplot(df['price'])
plt.title("Box plot de precios de productos")
plt.show()
```

## 3. Exploración de calificaciones de productos

```python
# Calcular medidas de centralidad
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

print(f"Media de calificaciones: {mean_rating}")
print(f"Mediana de calificaciones: {median_rating}")
print(f"Moda de calificaciones: {mode_rating}")

# Calcular medidas de dispersión
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Varianza de calificaciones: {variance_rating}")
print(f"Desviación estándar de calificaciones: {std_dev_rating}")
print(f"Rango intercuartílico de calificaciones: {iqr_rating}")

# Calcular sesgo (skewness) y curtosis (kurtosis)
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurt()

print(f"Sesgo: {skewness_rating}")
print(f"Curtosis: {kurtosis_rating}")

# Visualización: Histograma para calificaciones
plt.figure(figsize=(8,6))
sns.histplot(df['stars'], bins=50, kde=True)
plt.title("Distribución de calificaciones de productos")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```
